```python
import torch
from d2l import torch as d2l
```

```python
# 从pandas格式中提取NumPy格式，并将其转换为张量表示:
n_train = train_data.shape[0]
train_features = torch.tensor(all_features[:n_train].values,dtype=torch.float32) # float32减少空间
test_features = torch.tensor(all_features[n_train:].values,dtype=torch.float32)
train_labels = torch.tensor(train_labels.values.reshape(-1,1),dtype=torch.float32)

print(train_features.shape) # 两个维度，0维：样本数；1维：特征数
print(train_labels.shape) # 标签
```

## 建模

### MLP

参考：https://blog.csdn.net/weixin_45901519/article/details/118702141

```python
from torch import nn
```

```python
# 网路结构；三层
features = train_features.shape[1]
net = nn.Sequential(nn.Linear(features,250),
                   nn.ReLU(),
                   nn.Linear(250,150),
                   nn.ReLU(),
                   nn.Linear(150,1))
```

```python
#目标函数
loss = nn.MSELoss() # 1、方差 是各数据偏离平均值 差值的平方和 的平均数 2、均方误差（MSE）是各数据偏离真实值 差值的平方和 的平均数

# 更关心相对误差(y − y ^ )/y（RMSE）

def log_rmse(net,features,labels):
    clamp_preds = torch.clamp(net(features),1,float("inf"))  
    rmse = torch.sqrt(loss(torch.log(clamp_preds),torch.log(labels))) # RMSE一般指均方根误差
    return rmse.item()
```

### 训练

```python
def train_epoch(net,num_epochs,batch_size,weight_decay,learning_rate,train_features, train_labels, valid_features, valid_labels): # k-folds法将训练集进一步分为训练和验证
    # 构建数据集迭代器
    train_ls = []
    valid_ls = []
    train_iter = d2l.load_array((train_features,train_labels),batch_size)
    # 优化器
    optimizer = torch.optim.Adam(net.parameters(),lr=learning_rate,weight_decay = weight_decay) # 加正则
    for epoch in range(num_epochs):
        for X,y in train_iter:
            optimizer.zero_grad() # 梯度清除
            l = loss(net(X),y)
            l.backward()
            optimizer.step() # 更新每个参数
        # 每一个epoch保存训练和验证误差
        train_ls.append(log_rmse(net,train_features,train_labels))
        if valid_labels is not None:
            valid_ls.append(log_rmse(net,valid_features,valid_labels))
    return train_ls, valid_ls
```

### k折验证

```python
def get_k_fold_data(k, i, X, y):
    assert k > 1
    fold_size = X.shape[0] // k
    X_train, y_train = None, None
    for j in range(k):
        idx = slice(j * fold_size, (j + 1) * fold_size)
        X_part, y_part = X[idx, :], y[idx]
        if j == i:
            X_valid, y_valid = X_part, y_part
        elif X_train is None:
            X_train, y_train = X_part, y_part
        else:
            X_train = torch.cat([X_train, X_part], 0)
            y_train = torch.cat([y_train, y_part], 0)
    return X_train, y_train, X_valid, y_valid
```

```python
def k_fold_valid(k,net,X_train, y_train, num_epochs,batch_size,weight_decay,learning_rate):
    train_l_sum, valid_l_sum = 0, 0
    for i in range(k):
        data = get_k_fold_data(k, i, X_train, y_train)
        train_ls, valid_ls = train_epoch(net,num_epochs,batch_size,weight_decay,learning_rate,*data)
        train_l_sum += train_ls[-1] # 最后一个epoch（即收敛时候）的训练loss
        valid_l_sum += valid_ls[-1]
        if i == 0:
            d2l.plot(list(range(1, num_epochs + 1)), [train_ls, valid_ls],
                     xlabel='epoch', ylabel='rmse', xlim=[1, num_epochs],
                     legend=['train', 'valid'], yscale='log')
        print(f'折{i + 1}，训练log rmse{float(train_ls[-1]):f}, '
              f'验证log rmse{float(valid_ls[-1]):f}')
    return train_l_sum / k, valid_l_sum / k # k次均值
```

### 模型选择 

```python
# 参数在这里定义
learning_rate = 0.01
weight_decay = 0.1
batch_size = 64
k = 5
num_epochs = 50
```

```python
train_l, valid_l = k_fold_valid(k,net,train_features, train_labels, num_epochs,batch_size,weight_decay,learning_rate)
print(f'{k}-折验证: 平均训练log rmse: {float(train_l):f}, '
      f'平均验证log rmse: {float(valid_l):f}')
```

## 提交成绩

```python
def train_and_pred(net,train_features, test_features, train_labels, test_data,
                   num_epochs, lr, weight_decay, batch_size):
    train_ls, _ = train_epoch(net,num_epochs,batch_size,weight_decay,learning_rate,train_features,train_labels,None,None)
    d2l.plot(np.arange(1, num_epochs + 1), [train_ls], xlabel='epoch',
             ylabel='log rmse', xlim=[1, num_epochs], yscale='log')
    print(f'训练log rmse：{float(train_ls[-1]):f}')
    # 将网络应用于测试集。
    preds = net(test_features).detach().numpy()
    # 将其重新格式化以导出到Kaggle
    test_data['SalePrice'] = pd.Series(preds.reshape(1, -1)[0])
    # print(preds.shape) # (1459, 1)
    # print(preds.reshape(1, -1).shape) # (1, 1459)
    # print(preds.reshape(1, -1)) # [[119367.72 153963.94 198646.55 ... 208291.31 107010.71 240509.47]]
    submission = pd.concat([test_data['Id'], test_data['SalePrice']], axis=1)
    submission.to_csv('submission.csv', index=False)
```

```python
train_and_pred(net,train_features, test_features, train_labels, test_data,
                   num_epochs, learning_rate, weight_decay, batch_size)
```
